```python
def get_image_path(row):
    base = "/root/.cache/kagglehub/datasets/vikashrajluhaniwal/fashion-images/versions/1/data"

    category = row['Category']   # Apparel or Footwear
    gender = row['Gender']       # Boys, Girls, Men, Women
    image_file = row['Image']    # e.g. 42419.jpg

    path = os.path.join(base, category, gender, "Images", "images_with_product_ids", image_file)
    return path if os.path.exists(path) else None

# Apply to all rows
df['local_image_path'] = df.apply(get_image_path, axis=1)

# Check how many images were found
found = df['local_image_path'].notna().sum()
not_found = df['local_image_path'].isna().sum()

print(f"Images found: {found}")
print(f"Images NOT found: {not_found}")
print("\nSample paths:")
print(df['local_image_path'].dropna().head(3).tolist())
```

```python
from IPython.display import display, Image as IPImage
import random

# Show 3 random products with their details
samples = df.sample(3)

for _, row in samples.iterrows():
    print(f"Product: {row['ProductTitle']}")
    print(f"Colour: {row['Colour']} | Usage: {row['Usage']} | Gender: {row['Gender']}")
    print(f"Category: {row['Category']} > {row['SubCategory']} > {row['ProductType']}")
    display(IPImage(filename=row['local_image_path'], width=150))
    print("-"*50)
```

```python
# Create a rich text description for each product
# This is what Sentence Transformer will convert to vectors

def create_description(row):
    desc = f"{row['ProductTitle']} - {row['Colour']} {row['ProductType']} for {row['Gender']}. "
    desc += f"Category: {row['Category']}, {row['SubCategory']}. "
    desc += f"Suitable for {row['Usage']} use."
    return desc

df['description'] = df.apply(create_description, axis=1)

# Preview
print("Sample descriptions:")
for desc in df['description'].head(5).tolist():
    print("-", desc)
```
